```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, roc_auc_score
from imblearn.over_sampling import RandomOverSampler, SMOTE
from collections import Counter



# Read the CSV file
import io
df = pd.read_csv('aug_train.csv')

# Data preprocessing
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Handle missing values if any (for this dataset, it appears there are none)
# For demonstration, I'll proceed with the data as is

# Explore the target variable distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Response', data=df)
plt.title('Class Distribution Before Balancing')
plt.show()

# Separate features and target
X = df.drop(['id', 'Response'], axis=1)
y = df['Response']

# Convert categorical variables to numerical
X = pd.get_dummies(X, columns=['Gender', 'Vehicle_Age', 'Vehicle_Damage'], drop_first=True)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check class distribution
print("Original class distribution:", Counter(y_train))

# Function to evaluate models
def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Evaluate performance
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Initialize models
lr = LogisticRegression(max_iter=1000, random_state=42)
rf = RandomForestClassifier(random_state=42)

# Evaluate models on imbalanced data
print("\n" + "="*50)
print("Performance on Imbalanced Data")
print("="*50 + "\n")

print("\nLogistic Regression Performance:")
evaluate_model(lr, X_train, X_test, y_train, y_test)

print("\nRandom Forest Performance:")
evaluate_model(rf, X_train, X_test, y_train, y_test)

## Technique 1: Random Oversampling
print("\n" + "="*50)
print("Applying Random Oversampling")
print("="*50 + "\n")

ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(X_train, y_train)

print("Class distribution after Random Oversampling:", Counter(y_ros))

# Reinitialize models for oversampled data
lr_ros = LogisticRegression(max_iter=1000, random_state=42)
rf_ros = RandomForestClassifier(random_state=42)

# Evaluate models on oversampled data
print("\nLogistic Regression Performance with Random Oversampling:")
evaluate_model(lr_ros, X_ros, X_test, y_ros, y_test)

print("\nRandom Forest Performance with Random Oversampling:")
evaluate_model(rf_ros, X_ros, X_test, y_ros, y_test)

## Technique 2: SMOTE (Synthetic Minority Oversampling Technique)
print("\n" + "="*50)
print("Applying SMOTE")
print("="*50 + "\n")

smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train, y_train)

print("Class distribution after SMOTE:", Counter(y_smote))

# Reinitialize models for SMOTE data
lr_smote = LogisticRegression(max_iter=1000, random_state=42)
rf_smote = RandomForestClassifier(random_state=42)

# Evaluate models on SMOTE data
print("\nLogistic Regression Performance with SMOTE:")
evaluate_model(lr_smote, X_smote, X_test, y_smote, y_test)

print("\nRandom Forest Performance with SMOTE:")
evaluate_model(rf_smote, X_smote, X_test, y_smote, y_test)

# Create a comparison table of results
results = {
    'Model': ['Logistic Regression', 'Random Forest'],
    'Imbalanced Accuracy': [
        accuracy_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train).predict(X_test)),
        accuracy_score(y_test, RandomForestClassifier(random_state=42).fit(X_train, y_train).predict(X_test))
    ],
    'Imbalanced F1': [
        f1_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_train, y_train).predict(X_test)),
        f1_score(y_test, RandomForestClassifier(random_state=42).fit(X_train, y_train).predict(X_test))
    ],
    'ROS Accuracy': [
        accuracy_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_ros, y_ros).predict(X_test)),
        accuracy_score(y_test, RandomForestClassifier(random_state=42).fit(X_ros, y_ros).predict(X_test))
    ],
    'ROS F1': [
        f1_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_ros, y_ros).predict(X_test)),
        f1_score(y_test, RandomForestClassifier(random_state=42).fit(X_ros, y_ros).predict(X_test))
    ],
    'SMOTE Accuracy': [
        accuracy_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_smote, y_smote).predict(X_test)),
        accuracy_score(y_test, RandomForestClassifier(random_state=42).fit(X_smote, y_smote).predict(X_test))
    ],
    'SMOTE F1': [
        f1_score(y_test, LogisticRegression(max_iter=1000, random_state=42).fit(X_smote, y_smote).predict(X_test)),
        f1_score(y_test, RandomForestClassifier(random_state=42).fit(X_smote, y_smote).predict(X_test))
    ]
}

results_df = pd.DataFrame(results)
print("\nComparison of Results:")
print(results_df)

# Plot comparison
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(results_df['Model'], results_df['Imbalanced F1'], label='Imbalanced', marker='o')
plt.plot(results_df['Model'], results_df['ROS F1'], label='Random Oversampling', marker='o')
plt.plot(results_df['Model'], results_df['SMOTE F1'], label='SMOTE', marker='o')
plt.title('F1 Score Comparison')
plt.ylabel('F1 Score')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(results_df['Model'], results_df['Imbalanced Accuracy'], label='Imbalanced', marker='o')
plt.plot(results_df['Model'], results_df['ROS Accuracy'], label='Random Oversampling', marker='o')
plt.plot(results_df['Model'], results_df['SMOTE Accuracy'], label='SMOTE', marker='o')
plt.title('Accuracy Comparison')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
